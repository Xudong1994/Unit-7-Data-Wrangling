```python
# Read excel dataset
import pandas as pd
df = pd.read_csv("mobile_price_data.csv")
df
```

```python
df.info()
```

```python
df["mp_speed"].value_counts()
```

```python
# remove "₹" in mobile_price

df["mobile_price"] = df["mobile_price"].str.strip("₹")
df["mobile_price"] = df["mobile_price"].str.replace(",","")
df["mobile_price"] = df["mobile_price"].astype("int")
```

```python
# change dual_sim to 2, and single sim to 1
df["dual_sim"] = df["dual_sim"].str.replace("Dual Sim", "2")
df["dual_sim"] = df["dual_sim"].str.replace("Single Sim", "1")
df["dual_sim"] = df["dual_sim"].astype("int")
assert df["dual_sim"].dtype == "int"
df["dual_sim"].value_counts()
```

```python
# Remove GHz in mp_speed and transform to float type
df["mp_speed"] = df["mp_speed"].apply(lambda x:x[:-3])
```

```python
df["mp_speed"] = df["mp_speed"].astype("float")
```

```python
# Remove GB in memory and transform to int
df["int_memory"] = df["int_memory"].apply(lambda x:x[:-2])
df["int_memory"] = df["int_memory"].astype("int64")
```

```python
# # Remove GB in Ram and transform to int
df["ram"] = df["ram"].apply(lambda x:x[:-2])
df["ram"] = df["ram"].astype("int64")
```

```python
# # Remove mah in battery power and transform to int
df["battery_power"] = df["battery_power"].apply(lambda x:x[:-3])
df["battery_power"] = df["battery_power"].astype("int64")
```

```python
df.info()
```

```python
# Remove mm in mobile width and transform to float
df["mob_width"] = df["mob_width"].apply(lambda x:x[:-2])
df["mob_width"] = df["mob_width"].astype("float64")
```

```python
# Remove mm in mobile height and transform to float
df["mob_height"] = df["mob_height"].apply(lambda x:x[:-2])
df["mob_height"] = df["mob_height"].astype("float64")
```

```python
# Remove mm in mobile height and transform to float
df["mob_depth"] = df["mob_depth"].apply(lambda x:x[:-2])
df["mob_depth"] = df["mob_depth"].astype("float64")
```

```python
# Remove g in mobile weight and transform to float
df["mob_weight"] = df["mob_weight"].apply(lambda x:x[:-1])
df["mob_weight"] = df["mob_weight"].astype("float64")
```

```python
df["resolution"].value_counts()
```

```python
# Split resolution column
df["resolution"] = df["resolution"].str.replace("$", " ")
df["resolution"] = df["resolution"].str.replace("*", " x ")
df[["resolution_x", "x","resolution_y","Pixels"]] = df["resolution"].str.split(expand = True)
del df["x"]
del df["Pixels"]
```

```python
df["resolution"].value_counts()
```

```python
df[["disp_size_cm", "disp_size_inch"]] = df["disp_size"].str.split("(", expand = True)
```

```python
df
```

```python
df["disp_size_inch"].value_counts
```

```python
df["disp_size_inch"] = df["disp_size_inch"].apply(lambda x:x[:-5])
df["disp_size_inch"] = df["disp_size_inch"].astype("float64")
```

```python
df
```

```python
df["resolution_x"] = df["resolution_x"].astype("int64")
df["resolution_y"] = df["resolution_y"].astype("int64")
df["resolution_multiply"] = df["resolution_x"] * df["resolution_y"]
df = df[[*df.columns[:6], 'resolution_multiply', * df.columns[6:-1]]]
```

```python
df["num_cores"].value_counts()
```

```python
df["num_cores"] = df["num_cores"].map({"Octa Core" : 8, "Quad Core" : 4, "Single Core" : 1})
```

```python
df["os"].value_counts()
```

```python
df["disp_size_cm"] = df["disp_size_cm"].str.replace("cm", "")
df["disp_size_cm"] = df["disp_size_cm"].astype("float64")
```
